## **Handling Null Values**

```python
df[["Weapon Used Cd","Crm Cd 2", "Crm Cd 3", "Crm Cd 4"]] = df[["Weapon Used Cd", "Crm Cd 2", "Crm Cd 3", "Crm Cd 4"]].fillna(0)
```

Untuk 4 kolom diatas dilakukan penanganan khusus dalam pengisian NaN Values, yaitu dengan cara menggantinya menjadi 0. Alasannya karena 3 kolom diatas menjelaskan mengenai kasus kriminal lain selain kasus kriminal utama, karena itu tidak bisa diinput menggunakan metode statistik ataupun menggunakan tools lain. Untuk weapon used, diisi 0 juga karena bisa saja pembunuhan tidak dilakukan menggunakan senjata.

```python
df['Mocodes'].fillna('Unknown', inplace=True)
df['Vict Sex'].fillna('Not Specified', inplace=True)
df['Vict Descent'].fillna('Not Specified', inplace=True)
df['Premis Desc'].fillna('Not Specified', inplace=True)
df['Weapon Desc'].fillna('Not Specified', inplace=True)
df['Cross Street'].fillna('Unknown', inplace=True)
```

Untuk beberapa kolom diatas dilakukan penanganan yang berbeda juga:
1. Mocodes: Informasi tentang metode atau pola yang digunakan dalam melakukan kejahatan. Karena kolom ini bersifat penting, maka nilai NaN di isi dengan "Unknown"
2. Vict Sex: Jenis kelamin korban, diisi dengan "Not Specified".
3. Vict Descent: Etnis korban, diisi dengan "Not Specified".  
dll.

```python
most_frequent_status = df['Status'].mode()[0]
df['Status'].fillna(most_frequent_status, inplace=True)
```

```python
df['Premis Cd'].fillna(0, inplace=True)
df['Crm Cd 1'].fillna(0, inplace=True)
```

# **Visualisasi dan Insight**

## **Top 10 Jenis Kejahatan yang Paling Umum**

```python
crime_types = df['Crm Cd Desc'].value_counts()

plt.figure(figsize=(12,8))
crime_types[:10].plot(kind='barh', color='green')
plt.title('10 Jenis Kejahatan yang Paling Umum')
plt.xlabel('Jumlah Kejadian')
plt.ylabel('Jenis Kejahatan')
plt.show()
```

Dari grafik diatas dapat dilihat bahwa jenis kriminal yang sering terjadi adalah pencurian kendaraan. Ini menunjukkan bahwa pencurian kendaraan adalah masalah utama yang dihadapi dalam wilayah yang dianalisis. Mungkin beberapa alasan yang bisa menjadi penyebabnya adalah sebagai berikut:
1. Kendaraan memiliki nilai ekonomis tinggi, sehingga menjadi target yang menguntungkan bagi pelaku kriminal. Kendaraan curian bisa dijual kembali secara utuh atau dipreteli dan dijual dalam bentuk suku cadang, yang seringkali lebih sulit dilacak.
2. Kendaraan adalah aset bergerak yang tersebar di banyak tempat, terutama di kota-kota besar. Banyak kendaraan diparkir di area umum seperti tempat parkir, jalanan, atau bahkan area perumahan tanpa pengawasan ketat, sehingga membuatnya lebih rentan dicuri.
3. Banyak pencurian kendaraan merupakan bagian dari kejahatan terorganisir yang melibatkan jaringan penjualan mobil curian, baik secara lokal maupun internasional. Sindikat kriminal memiliki keterampilan khusus dan strategi untuk mencuri kendaraan dalam jumlah besar dan menjualnya secara ilegal.

## **Pencurian Kendaraan Berdasarkan Wilayah**

```python
vehicle_theft = df[df['Crm Cd Desc'] == 'VEHICLE - STOLEN']
vehicle_theft_by_area = vehicle_theft['AREA NAME'].value_counts().nlargest(10)

plt.figure(figsize=(10,5))
vehicle_theft_by_area.plot(kind='barh', color='orange')
plt.xlabel('Jumlah Pencurian Kendaraan')
plt.ylabel('Area')
plt.title('10 Area dengan Pencurian Kendaraan Tertinggi')
plt.gca().invert_yaxis()
plt.show()
```

Beberapa insight yang didapatkan:
1. Berdasarkan konsentrasi pencurian kendaraan yang tinggi di area tertentu, terutama 77th Street, ini bisa menjadi sinyal perlunya alokasi sumber daya keamanan yang lebih efektif. Alih-alih menyebar personel keamanan secara merata, pihak berwenang dapat memprioritaskan area-area dengan risiko tinggi. Hal ini bisa mencakup peningkatan patroli di jam-jam rawan atau penerapan teknologi berbasis kecerdasan buatan untuk mendeteksi aktivitas mencurigakan di area-area tersebut.
2. Area seperti 77th Street dan Newton mungkin membutuhkan pendekatan pencegahan yang lebih holistik. Selain penegakan hukum, pemberdayaan komunitas dengan melibatkan penduduk lokal dalam program pencegahan kejahatan, seperti neighborhood watch atau pemberdayaan ekonomi, bisa menjadi solusi jangka panjang.
3. Ada kemungkinan area-area dengan pencurian kendaraan tinggi memiliki faktor-faktor infrastruktur yang mendukung kegiatan kriminal, seperti parkir terbuka yang tidak diawasi atau pencahayaan yang buruk di jalanan. Melakukan audit infrastruktur dan memperbaiki kondisi seperti ini bisa mengurangi peluang kejahatan.

## **Pola Musiman (Bulan Terjadinya Pencurian Kendaraan)**

```python
vehicle_theft['DATE OCC'] = pd.to_datetime(vehicle_theft['DATE OCC'])
vehicle_theft['Month'] = vehicle_theft['DATE OCC'].dt.month
vehicle_theft_by_month = vehicle_theft['Month'].value_counts().sort_index()

# Visualisasi: Pola Musiman Pencurian Kendaraan
plt.figure(figsize=(10,5))
vehicle_theft_by_month.plot(kind='bar', color='blue')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Pencurian Kendaraan')
plt.title('Pola Musiman Pencurian Kendaraan')
plt.show()
```

Berdasarkan grafik diatas, terlihat ada sedikit penurunan pencurian kendaraan pada bulan September hingga Desember. Maka dari itu, penegak hukum dapat menggunakan periode ini untuk fokus pada upaya pencegahan yang lebih proaktif, seperti kampanye kesadaran publik, pemasangan CCTV tambahan, atau peningkatan patroli pada bulan-bulan dengan angka pencurian yang lebih tinggi, seperti Juli dan Agustus. Selain itu ada beberapa rekomendasi/saran yang boleh digunakan oleh pemerintah daerah:
- Bulan-bulan tertentu, seperti awal dan pertengahan tahun, bisa menjadi waktu yang tepat untuk mengedukasi masyarakat terkait langkah-langkah pencegahan pencurian kendaraan, seperti penggunaan kunci pengaman tambahan atau parkir di lokasi yang lebih aman.
- Peningkatan pencurian pada bulan-bulan musim panas (seperti Juli dan Agustus) mungkin mengindikasikan perubahan perilaku pelaku kejahatan berdasarkan cuaca, kegiatan outdoor yang meningkat, atau liburan sekolah.
