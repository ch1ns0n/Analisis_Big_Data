## **Visualisasikan event secata spatial yang dipilih dalam bentuk proyeksi peta menggunakan Geo Pandas atau yang lain**

```python
# Filter EventCode 111 untuk event di Indonesia
df_111 = df_cleaned[df_cleaned['EventCode'] == 111]

# Handle missing data (drop rows with missing coordinates)
df_111 = df_111.dropna(subset=['Actor1Geo_Long', 'Actor1Geo_Lat'])

# Buat GeoDataFrame dengan kolom longitude dan latitude
geometry = [Point(xy) for xy in zip(df_111['Actor1Geo_Long'], df_111['Actor1Geo_Lat'])]
gdf_111 = gpd.GeoDataFrame(df_111, geometry=geometry)

# Atur sistem koordinat CRS (Coordinate Reference System) ke WGS84
gdf_111.crs = "EPSG:4326"

# Pastikan untuk menyesuaikan path ke lokasi file yang sudah diekstrak
world = gpd.read_file("ne_110m_admin_0_countries.shp")

# Koordinat bounding box Indonesia
minx, miny, maxx, maxy = 95, -11, 141, 6  # Sekitar koordinat Indonesia

# Plotting peta
fig, ax = plt.subplots(figsize=(15, 10))

# Potong peta dunia hanya untuk wilayah Indonesia
world_indonesia = world.cx[minx:maxx, miny:maxy]

# Plot data peta dunia di sekitar Indonesia
world_indonesia.plot(ax=ax, color='lightgrey', edgecolor='black')

# Buat kolom baru untuk ukuran marker
gdf_111['size'] = np.log(gdf_111.shape[0]) * 5  # Menyesuaikan ukuran marker

# Plot event dengan EventCode 111
sc = gdf_111.plot(ax=ax, color='red', markersize=gdf_111['size'], alpha=0.7, marker='o', label='EventCode 111')

# Plot beberapa kota besar di Indonesia (menambahkan informasi geografis)
cities = {
    "Jakarta": (106.8456, -6.2088),
    "Surabaya": (112.7508, -7.2575),
    "Medan": (98.6722, 3.5897),
    "Denpasar": (115.2167, -8.6500),
}

for city, coord in cities.items():
    plt.text(coord[0], coord[1], city, fontsize=9, ha='right', color='blue', fontweight='bold')

# Fokus pada wilayah Indonesia dengan bounding box
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)

# Tambahkan title dan label
plt.title('Visualisasi Geospasial Event dengan EventCode 111', fontsize=18, fontweight='bold')
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)

# Tambahkan legenda
plt.legend(loc='upper left')

# Tambahkan grid
ax.grid(color='grey', linestyle='--', linewidth=0.5)

# Tampilkan peta
plt.tight_layout()
plt.show()
```

## **Visualisasi jumlah event secara temporal menggunakan visualisasi yang cocok**

```python
# Mengubah kolom 'SQLDATE' menjadi tipe datetime
df_111['Date'] = pd.to_datetime(df_111['SQLDATE'], format='%Y%m%d')

# Agregasi jumlah event per bulan
event_counts_monthly = df_111.groupby(df_111['Date'].dt.to_period('M')).size()

# Visualisasi jumlah event per bulan
plt.figure(figsize=(12, 6))
event_counts_monthly.plot(kind='line', marker='o', color='cornflowerblue', linewidth=2)
plt.title('Jumlah Event per Bulan (EventCode 111)', fontsize=15)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Jumlah Event', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## **Visualisasikan Rata-Rata Tone Event yang diplih (AvgTone) secara temporal menggunakan visualisasi yang cocok**

```python
# Ubah kolom 'SQLDATE' menjadi tipe datetime
df_cleaned['Date'] = pd.to_datetime(df_cleaned['SQLDATE'], format='%Y%m%d')

# Agregasi rata-rata tone event per hari
avg_tone_per_day = df_cleaned.groupby(df_cleaned['Date'].dt.date)['AvgTone'].mean()

# Membuat DataFrame dari hasil agregasi
avg_tone_df = pd.DataFrame(avg_tone_per_day).reset_index()
avg_tone_df.columns = ['Date', 'Avg_Tone']

# Plotting rata-rata tone secara temporal
plt.figure(figsize=(12, 6))
plt.plot(avg_tone_df['Date'], avg_tone_df['Avg_Tone'], marker='o', linestyle='-', color='orange')
plt.title('Rata-Rata Tone Event Secara Temporal', fontsize=18, fontweight='bold')
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Rata-Rata Tone', fontsize=12)
plt.xticks(rotation=45)
plt.grid(color='grey', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
