# **Content-Based Filtering**

**Vectorizing Features**

```python
# Identifikasi kolom deskriptif (misalnya semua kecuali 'item_id')
descriptive_columns = item_fea.columns.difference(['item_id'])

# Gabungkan semua kolom deskriptif menjadi satu string per baris
item_fea['combined_features'] = item_fea[descriptive_columns].astype(str).apply(lambda row: ' '.join(row), axis=1)
```

```python
# Vectorize features
vectorizer = TfidfVectorizer()
item_vectors = vectorizer.fit_transform(item_fea['combined_features'])
```

**Menghitung kemiripan antar item menggunakan cosine similarity**

```python
# Compute similarity matrix
item_similarity = cosine_similarity(item_vectors)

item_id = 0
similar_items = list(enumerate(item_similarity[item_id]))
similar_items = sorted(similar_items, key=lambda x: x[1], reverse=True)
```

```python
# Tampilkan 5 item paling mirip
print("5 Item paling mirip dengan Item ID:", item_id)
for sim_item in similar_items[:5]:
    print(f"Item ID: {sim_item[0]}, Similarity: {sim_item[1]}")
```

**Prediksi Rating**

```python
def predict_rating_content_based(item_id, item_similarity, train_data):
    similar_items = item_similarity[item_id].argsort()[-5:]  # Top 5 similar items
    ratings = train_data.loc[train_data['item_id'].isin(similar_items), 'rating']
    return ratings.mean() if not ratings.empty else train_data['rating'].mean()
```

```python
test_data['predicted_rating_cbf'] = test_data['item_id'].apply(lambda x: predict_rating_content_based(x, item_similarity, train_data))
```

# **Collaborative Filtering**

```python
# Siapkan data untuk Surprise library
reader = Reader(rating_scale=(0, 5))
data = Dataset.load_from_df(train_data[['user_id', 'item_id', 'rating']], reader)

# Split data
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
# Train SVD model
model = SVD()
model.fit(trainset)

# Prediksi rating untuk test_data
test_data['predicted_rating_cf'] = test_data.apply(
    lambda row: model.predict(row['user_id'], row['item_id']).est, axis=1
)
```
